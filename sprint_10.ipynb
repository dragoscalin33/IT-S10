##6. Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any.

````python
# se hace asi: df['Data_naixement'] = pd.to_datetime(df[['any', 'mes', 'dia']])
# pero python solo lo reconoce si los nombres de las columnas son: 'year', 'month', 'day'

# asi que llamamos la columna y la convertemos en str, el errors='coerce' es por si encuentra un error, le decimos que devuelva NaT (not a time)
df['Data_naixement'] = pd.to_datetime(df['Dia'].astype(str) + '-' +
                                      df['Mes'].astype(str) + '-' +
                                      df['Any'].astype(str),
                                      errors='coerce')
df
````

````python
from datetime import date

def calcular_edat(data_naixement):
    if pd.isna(data_naixement):  # Si la fecha no existe (NaT)
        return None

    avui = date.today()
    # si todavia no ha llegado el dia del cumpleaños, la comparacion te devuelve True, que es igual a 1, así te devuelve la edad bien, le faltan unos dias todavia para el cumple de este año
    return avui.year - data_naixement.year - ((avui.month, avui.day) < (data_naixement.month, data_naixement.day)) # si te devuelve True

df['Edat'] = df['Data_naixement'].apply(calcular_edat)
df
````

## 1. Crea una funció que prengui un dataframe com a paràmetre d'entrada.

* La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:
```
un histograma/boxplot si la variable és numèrica
unes barres dels valors més freqüents si és categòrica
unes barres dels anys més freqüents si la dada està en format data.
```
* La idea és crear una funció que funcioni per **qualsevol** dataframe, no només amb el que hem treballat fins ara.

* Mostra el resultat de la funció en algun dels datasets d’exemple que conté el paquet seaborn. Per exemple, iris, penguins o titanic.

**Tingues en consideració que en el següent sprint treballaràs exclusivament amb gràfics. L’objectiu d’aquest exercici no és crear gràfics molt elaborats, sinó resoldre una necessitat de manera ràpida i automàtica.**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Ya importado, pero por si acaso
from pandas.api.types import is_numeric_dtype, is_datetime64_any_dtype, is_object_dtype

# 1. Definimos la función
def generar_graficos_auto(dataframe, path_prefix="grafico_"):
    """
    Recorre cada columna de un DataFrame y genera y guarda un gráfico apropiado
    para su tipo de dato.
    """

    # Iteramos sobre cada nombre de columna
    for col in dataframe.columns:
        print(f"Procesando columna: {col}")

        # Creamos una nueva figura para cada gráfico
        plt.figure(figsize=(10, 6))

        try:
            # Opción 1: Si la columna es NUMÉRICA
            if is_numeric_dtype(dataframe[col]):
                sns.histplot(dataframe[col], kde=True, bins=30)
                plt.title(f"Histograma de {col}")
                plt.xlabel(col)
                plt.ylabel("Frecuencia")

            # Opción 2: Si la columna es de tipo FECHA/HORA
            elif is_datetime64_any_dtype(dataframe[col]):
                # Extraemos el año y contamos los valores
                anyos = dataframe[col].dt.year.value_counts().sort_index()
                sns.barplot(x=anyos.index, y=anyos.values)
                plt.title(f"Recuento por Año en {col}")
                plt.xlabel("Año")
                plt.ylabel("Recuento")

            # Opción 3: Si es CATEGÓRICA (object, string, category)
            # O si tiene pocos valores únicos (puede ser numérica pero categórica, como 'Pclass' 1, 2, 3)
            elif is_object_dtype(dataframe[col]) or dataframe[col].nunique() < 20:
                # Contamos los valores más frecuentes (top 15)
                top_15_categorias = dataframe[col].value_counts().nlargest(15).index
                sns.countplot(data=dataframe, x=col, order=top_15_categorias)
                plt.title(f"Recuento de valores más frecuentes en {col}")
                plt.xlabel(col)
                plt.ylabel("Recuento")
                plt.xticks(rotation=45) # Giramos las etiquetas por si son largas

            # Si no cumple nada, lo saltamos
            else:
                print(f"   -> Columna {col} saltada (tipo no manejado o demasiadas categorías).")
                plt.close() # Cerramos la figura vacía
                continue # Pasamos a la siguiente iteración del bucle

            # 3. Guardamos el gráfico en un fichero
            nombre_archivo_grafico = f"{path_prefix}{col}.png"
            plt.tight_layout() # Ajusta el gráfico para evitar cortes
            plt.savefig(nombre_archivo_grafico)
            print(f"   -> Gráfico guardado como: {nombre_archivo_grafico}")

        except Exception as e:
            print(f"   -> ERROR al procesar {col}: {e}")

        finally:
            plt.close() # Cerramos la figura para liberar memoria (muy importante)

# 2. Probamos la función con el dataset 'titanic' de Seaborn
print("\n--- Probando con el dataset 'titanic' ---")
titanic = sns.load_dataset('titanic')
generar_graficos_auto(titanic, path_prefix="titanic_")

print("\n--- Probando con nuestro dataset 'df_actualizado' ---")
# (Nos aseguramos de que Data_naixement sea datetime)
df_actualizado['Data_naixement'] = pd.to_datetime(df_actualizado['Data_naixement'])
generar_graficos_auto(df_actualizado, path_prefix="trabajadores_")
````

##2. Carrega l'arxiu matriu_distancies.xlsx a pandas, de manera que els noms de files i els noms de columnes siguin els de les ciutats. Borra "Las Palmas de Gran Canaria" i "Palma" perquè poguem fer el trajecte en cotxe.

Font: Mejores Rutas

* Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible.

* No cal que ho facis de forma òptima, ens interessa que desenvolupis una solució raonable utilitzant les eines que tens actualment.

* Per exemple, una aproximació senzilla (que no òptima) seria anant sempre a la ciutat més propera que no haguem visitat encara

* Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

* EXTRA: Des de quina ciutat la ruta seria més curta amb l'algoritme plantejat

````python
import pandas as pd
import numpy as np

# 1. Cargar el fichero Excel
print("Cargando matriz...")
try:
    matriz_distancias = pd.read_excel("/content/matriu_distancies.xlsx", index_col=0)
except FileNotFoundError:
    print("Error: No se encontró el archivo '/content/matriu_distancies.xlsx'")
    print("Por favor, asegúrate de que el archivo está subido a Colab.")
    # Si da error, usa esta línea y sube el CSV que me mostraste
    # matriz_distancias = pd.read_csv("/content/matriu_distancies.xlsx - Distancia (kilómetros).csv", index_col=0)

# 2. Ciudades a borrar
ciudades_a_borrar = ["Las Palmas de Gran Canaria", "Palma"]

# 3. Borrar de filas y columnas
matriz_limpia = matriz_distancias.drop(index=ciudades_a_borrar, columns=ciudades_a_borrar, errors='ignore')

# 4. SOLUCIÓN AL ERROR DE 'MADRID': Limpiamos espacios
matriz_limpia.index = matriz_limpia.index.str.strip()
matriz_limpia.columns = matriz_limpia.columns.str.strip()

# 5. SOLUCIÓN AL ERROR DE '2778 km': Forzamos que todo sea numérico
for col in matriz_limpia.columns:
    if matriz_limpia[col].dtype == 'object':
         matriz_limpia[col] = matriz_limpia[col].astype(str).str.replace('.', '', regex=False)

    matriz_limpia[col] = pd.to_numeric(matriz_limpia[col], errors='coerce')

# 6. Rellenamos cualquier NaN (incluida la diagonal) con 0
matriz_limpia = matriz_limpia.fillna(0)

print("Matriz de distancias limpia y convertida a números:")
print(matriz_limpia.info())
````

````python
# Definimos la función del algoritmo
def calcular_ruta_corta(matriz, ciudad_origen):
    ciudades_a_visitar = list(matriz.index)

    if ciudad_origen not in ciudades_a_visitar:
        return f"Error: La ciudad '{ciudad_origen}' no existe en la matriz.", 0

    ruta = [ciudad_origen]
    ciudades_a_visitar.remove(ciudad_origen)
    distancia_total = 0
    ciudad_actual = ciudad_origen

    while ciudades_a_visitar:
        distancias = matriz[ciudad_actual]
        distancias_restantes = distancias[ciudades_a_visitar]

        # Filtramos distancias > 0 para no quedarnos atascados
        distancias_sin_cero = distancias_restantes[distancias_restantes > 0]

        if not distancias_sin_cero.empty:
            distancia_viaje = distancias_sin_cero.min()
            ciudad_proxima = distancias_sin_cero.idxmin()
        else:
            # Si solo quedan ciudades a 0km (o es la última)
            distancia_viaje = distancias_restantes.min()
            ciudad_proxima = distancias_restantes.idxmin()

        distancia_total += distancia_viaje
        ciudad_actual = ciudad_proxima
        ruta.append(ciudad_actual)
        ciudades_a_visitar.remove(ciudad_actual)

    return ruta, distancia_total
````

````python
# Probamos la función empezando desde Madrid
ruta_madrid, dist_madrid = calcular_ruta_corta(matriz_limpia, 'Madrid')

if isinstance(ruta_madrid, list):
    print(f"Ruta comenzando en Madrid ({len(ruta_madrid)} ciudades):")
    print(" -> ".join(ruta_madrid))
    print(f"\nDistancia total: {dist_madrid:.0f} km")
else:
    print(ruta_madrid) # Imprimirá el error si lo hay
````

````python
print("\n" + "-" * 30)
print("EXTRA: Buscando la mejor ciudad de origen...")

resultados = {}
mejor_distancia = float('inf') # 'inf' es un número infinitamente grande
mejor_ciudad = None

for ciudad in matriz_limpia.index:
    ruta, distancia = calcular_ruta_corta(matriz_limpia, ciudad)
    resultados[ciudad] = {'ruta': ruta, 'distancia': distancia}

    if distancia < mejor_distancia:
        mejor_distancia = distancia
        mejor_ciudad = ciudad

print("\n🏆 ¡MEJOR RUTA ENCONTRADA!")
print(f"Ciudad de origen: {mejor_ciudad}")
print(f"Distancia total: {mejor_distancia:.0f} km")
print("Ruta:")
print(" -> ".join(resultados[mejor_ciudad]['ruta']))
````
